```python
import os

def merge_txt_files(input_folder, output_file):
    with open(output_file, 'w') as output:
        # read up
        txt_files = [file for file in os.listdir(input_folder) if file.endswith('.txt')]
        
        # write in
        for txt_file in txt_files:
            file_path = os.path.join(input_folder, txt_file)
            with open(file_path, 'r') as input_file:
                output.write(input_file.read())
                output.write('\n')  # create new line when reading next txt file.

# input and output files
input_folder = '/kaggle/input/cma-best-track-data/CMABSTdata'
output_file = '/kaggle/working/merged_txt_file.txt'

# Merge the txt files
merge_txt_files(input_folder, output_file)
```

```python
def reformat_data(input_file, output_file):
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        current_header = None

        for line in infile:
            if line.startswith('66666'):
                current_header = line.strip()
            else:
                formatted_line = f"{current_header} {line.strip()}\n"
                outfile.write(formatted_line)


input_file = '/kaggle/working/merged_txt_file.txt'
output_file = '/kaggle/working/cleaned.txt'


reformat_data(input_file, output_file)
```

Transform the txt file into CSV.

```python
import pandas as pd

def txt_to_csv(input_file, output_csv):
    # read
    with open(input_file, 'r') as infile:
        lines = infile.readlines()

    # df
    data = []
    for line in lines:
        line = line.strip().split()
        data.append(line)

    columns = [
        'Identifier', 'International Number ID', 'Lines of Data', 'Tropical Cyclone Number',
        'Chinese Tropical Cyclone Data', 'Tropical Cyclone End Record', 'Number of Hours between Paths per Line',
        'Name', 'Date of Record', 'Time', 'Grade', 'Latitude', 'Longitude',
        'Minimum Central Pressure', 'Maximum Wind Speed', 'Average Wind Speed'
    ]

    df = pd.DataFrame(data, columns=columns)
    df = df.drop(['Identifier', 'Lines of Data'], axis=1)
    df = df.dropna(subset=columns[7:14])
    # csv
    df.to_csv(output_csv, index=False)


input_txt = '/kaggle/working/cleaned.txt'
output_csv = '/kaggle/working/CMA_Best_Track_Data.csv'

txt_to_csv(input_txt, output_csv)
```
